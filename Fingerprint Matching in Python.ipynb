```python
import os
import cv2

# Charger l'image d'échantillon
sample = cv2.imread("SOCOFing/Altered/Altered-Hard/150__M_Right_index_finger_Obl.BMP")

# Initialiser les variables
best_score = 0
filename = None
image = None
kp1, kp2, mp = None, None, None
counter = 0

# Parcourir les 1000 premiers fichiers dans le répertoire "SOCOFing/Real"
for file in os.listdir("SOCOFing/Real")[:1000]:
    if counter % 10 == 0:
        print(counter)
        print(file)
    
    counter += 1
    # Charger l'image d'empreinte digitale actuelle
    fingerprint_image = cv2.imread("SOCOFing/Real/" + file)
    
    # Vérifier si l'image a été chargée correctement
    if fingerprint_image is not None:
        # Extraire les points clés avec l'algorithme SIFT
        sift = cv2.SIFT_create()
        keypoints_1, descriptors_1 = sift.detectAndCompute(sample, None)
        keypoints_2, descriptors_2 = sift.detectAndCompute(fingerprint_image, None)
        
        # Faire correspondre les descripteurs entre les deux images
        matches = cv2.FlannBasedMatcher({"algorithm": 1, 'trees': 10},
                                         {}).knnMatch(descriptors_1, descriptors_2, k=2)
        
        # Filtrer les correspondances en fonction de la distance
        match_points = [p for p, q in matches if p.distance < 0.1 * q.distance]
        
        # Calculer le nombre de points clés
        keypoints = min(len(keypoints_1), len(keypoints_2))
        
        # Calculer le score de correspondance
        current_score = len(match_points) / keypoints * 100
        
        # Mettre à jour les variables si le score actuel est meilleur
        if current_score > best_score:
            best_score = current_score
            filename = file
            image = fingerprint_image
            kp1, kp2, mp = keypoints_1, keypoints_2, match_points

# Afficher les résultats
print("The Best Match: " + str(filename))
print("Score: " + str(best_score))

# Dessiner les correspondances sur l'image résultante
result = cv2.drawMatches(sample, kp1, image, kp2, mp, None)

# Redimensionner l'image résultante
result = cv2.resize(result, None, fx=4, fy=4)

# Afficher l'image résultante
cv2.imshow("Result: ", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
